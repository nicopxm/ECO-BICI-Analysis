# Summary of Findings for Alcaldia Coyoacan
## Trip Counts and Stations:

### - Coyoacan has fewer trips and stations compared to previous alcaldias.
### - Colonia Del Carmen shows significantly higher usage and preference than all other colonias.
## Trip Duration:

### - Villa Coyoacan and Parque San Andres have the highest trip durations (18.5 min) but only 2-3 stations.
### - Del Carmen has a lower trip duration (16 min).
### - Stations with high trip duration are located at the boundaries and some in the center of Coyoacan.
## Trip Counts by Month:

### - Del Carmen dominates bike usage monthly.
### - Other colonias like Santa Catarina, San Mateo, and Parque San Andres have similar trip counts.
### - April recorded the highest number of trips in Del Carmen (23.5k).
## Trip Duration by Month:

### - Parque San Andres has the highest average trip duration in most months (20 min).
### - July-September show the highest trip duration for Parque San Andres.
### - San Diego Churubusco recorded the highest trip duration in January (19.9 min).
### - Villa Coyoacan was the highest in February and November.
## Demographics:

### - Most trips are taken by the 26-35 age group.
### - The 16-25 age group has higher trip counts than the 36-45 age group in Del Carmen, San Mateo, Barrio de Santa Catarina, and Barrio La Concepcion.
### - The 61-99 age group leads in trip duration in several colonias, with Barrio de Santa Catarina showing the highest (20.7 min).
## Gender Insights:

### - Males lead in trip duration in Parque San Andres (18.6 min).
### - Females lead in other colonias, with Villa Coyoacan recording nearly 20 min.
### - Barrio San Luca has the highest percentage of females (34%), followed by Villa Coyoacan (33.6%) and Del Carmen (30%).
### - High female percentages are noted for the 16-25 and 61-99 age groups in Barrio San Luca.
## Daily Activity:

### - Del Carmen leads in trip count every day, with fewer trips on weekends.
### - Trip duration is highest on Sundays, with Parque San Andres recording 22 min and Villa Coyoacan 20.6 min.
### - Males have more trips than females in every colonia.
## Age Group Insights by Day:

### - 26-35 age group leads weekday trip counts.
### - 61-99 age group leads weekend trip duration.
### - 16-25 age group records the lowest trip durations daily.
### - Highest percentages for the 61-99 age group are seen in San Diego Churubusco (Saturday) and Barrio San Luca (Sunday).
### - The 46-60 age group shows the highest percentages on Tuesday in Barrio San Luca.
### - 36-45 age group peaks on Saturday in San Diego Churubusco and Barrio San Luca.
## Hourly Data:

### - Del Carmen leads in hourly trip counts, with peaks at 9am, 2pm, and 6pm.
### - Top 3 colonias for females hourly are Del Carmen, Santa Catarina, and Parque San Andres.
### - Del Carmen also leads in hourly trips for the 16-25, 26-35, 36-45, 46-60, and 61-99 age groups.
### - Targeted Strategies
## Increase Usage in Underutilized Colonias:

### - Focus marketing campaigns and promotions in colonias with lower trip counts such as Santa Catarina, San Mateo, and Parque San Andres.
### - Enhance bike infrastructure and safety features in these areas to attract more users.
## Target Age Groups Effectively:

### - 16-25 Age Group: Increase promotional activities and events in Del Carmen, San Mateo, Barrio de Santa Catarina, and Barrio La Concepcion.
### - 26-35 Age Group: Promote bike-sharing during peak hours (8am, 2pm, 6pm) and weekdays to capitalize on their high usage.
### - 61-99 Age Group: Create senior-friendly programs and incentives in Barrio de Santa Catarina, Barrio San Luca, and Barrio La Concepcion, focusing on weekends.
## Gender-Specific Initiatives:

### - Females: Enhance safety and visibility of bike stations, especially in Villa Coyoacan, Del Carmen, and Barrio San Luca. Offer female-centric promotions and events.
### - Males: Encourage more bike usage in Parque San Andres and other areas where males lead in trip duration through competitive challenges or rewards.
## Enhance Weekend and Hourly Usage:

### - Promote leisure biking activities and events in Parque San Andres and Villa Coyoacan on weekends.
### - Introduce time-based promotions to encourage biking during peak hours (9am, 2pm, 6pm).
## Station Placement and Accessibility:

### - Optimize station placement by adding more stations in high trip duration areas at the boundaries and center of Coyoacan.
### - Ensure easy access to stations in high usage areas like Del Carmen and along main streets.
### - By implementing these targeted strategies, Coyoacan can enhance bike-sharing usage across various demographics and improve overall service efficiency and user satisfaction.
